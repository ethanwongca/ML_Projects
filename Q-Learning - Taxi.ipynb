## Q-Learning how to drive a Taxi
### Imports

```python
import numpy as np
import gym, random
```

## Creating the environment

```python
env = gym.make("Taxi-v2")
action_size = env.action_space.n
# sometimes this works!
state_size = env.observation_space.n
qtable = np.zeros((state_size, action_size))
```

## Bellman eq. and training setup

```python
total_episodes = 50000
total_test_episodes = 100
max_steps = 99
learning_rate = 0.7
gamma = 0.618
epsilon = 1.0
max_epsilon = 1.0
min_epsilon = 0.01
decay_rate = 0.01
```

```python

for episode in range(total_episodes):
    prev_state = env.reset()
    step = 0
    done = False
    
    for step in range(max_steps):
        exp_exp_tradeoff = random.uniform(0,1)

        if exp_exp_tradeoff > epsilon:
            action = np.argmax(qtable[prev_state,:])
        else:
            action = env.action_space.sample()
        new_state, reward, done, info = env.step(action)
        # the bellman equation!
        qtable[prev_state, action] = qtable[prev_state, action] + learning_rate * (reward + gamma * 
                                                                        np.max(qtable[new_state, :]) 
                                                                        - qtable[prev_state, action])
        prev_state = new_state
        if done:
            break
    epsilon = min_epsilon + (max_epsilon - min_epsilon) * np.exp(-decay_rate * episode)
```

```python
env.reset()
rewards = []
for episode in range(total_test_episodes):
    prev_state = env.reset()
    step = 0
    done = False
    total_rewards = 0
    for step in range(max_steps):
        env.render()
        action = np.argmax(qtable[prev_state,:])
        new_state, reward, done, info = env.step(action)
        total_rewards += reward
        if done:
            rewards.append(total_rewards)
            break
        prev_state = new_state
        print("Score over time: " + str(sum(rewards)/total_test_episodes))
    env.close()
```
